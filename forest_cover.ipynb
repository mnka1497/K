```python
import pandas as pd
import csv
```

```python
st = "C:\\Users\\XLMR3003\\Downloads\\learn-together\\train.csv"
df = pd.read_csv(st,encoding = "ISO-8859-1" ,error_bad_lines=False)
```

```python
df.head()
```

```python
df.dropna(axis=0, how='any', thresh=None, subset=None, inplace=True)
df.reset_index(drop=True, inplace=True)
df = df.drop(['Id'], axis= 1)
```

```python
df.head()
```

```python
X=df.iloc[:,:-1]
X.head()
```

```python
y = df['Cover_Type']
y.head()
```

```python
import sklearn
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 2)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 0)
classifier.fit(X_train, y_train)

# Predicting the Test set results

y_pred = classifier.predict(X_test)
report = classification_report(y_test, y_pred)
print(report)
```

```python
st = "C:\\Users\\XLMR3003\\Downloads\\learn-together\\test.csv"
df1 = pd.read_csv(st,encoding = "ISO-8859-1" ,error_bad_lines=False)
test_ids = df1["Id"]
df1=df1.drop(['Id'], axis =1)
```

```python
X_testing = sc.fit_transform(df1)
y_pred_test = classifier.predict(X_testing)
```

```python
output = pd.DataFrame({'ID': test_ids,
                       'TARGET': y_pred_test})
output.to_csv('submission.csv', index=False)
```

```python
param_grid = { 
    'n_estimators': [65,70, 80, 90,100],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [4,5,6,7,8],
    'criterion' :['gini', 'entropy']
}
```

```python
from sklearn.model_selection import GridSearchCV
CV_rfc = GridSearchCV(estimator=classifier, param_grid=param_grid, cv= 5)
CV_rfc.fit(X_train, y_train)
```
